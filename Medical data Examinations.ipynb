### D- Column wrangling.:

Let's create the columns we need for our analysis: BMI, cholesterolN, and glucN.

* Let's build the overweight column:

```python
df['BMI'] = df['weight']/(df['height']/100)**2
df['BMI']
```

```python
df['overweight'] = df['BMI'].apply(lambda x:1 if x>25 else 0)
df['overweight']
```

```python
df
```

* Let's build the **cholesterolN** column:

```python
df['cholesterolN'] = df['cholesterol'].apply(lambda x: 0 if x == 1 else 1)
df
```

* Let's build the **glucN** column:

```python
df['glucN'] = df['gluc'].apply(lambda x: 0 if x == 1 else 1)
df
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

### E- Analysis and visualization:

Let's Create a DataFrame for cat plot using `pd.melt` using just the values from 'overweight','cholesterolN', 'glucN', 'smoke', 'alco', and 'active'.

```python
df_cat = pd.melt(df, id_vars=['cardio'], value_vars=['cholesterolN', 'glucN', 'smoke', 'alco', 'active', 'overweight'])
df_cat
```

Let's Group and reformat the data to split it by 'cardio'. It Shows the counts of each feature. We will have to rename one of the columns for the catplot to work correctly.

```python
df_cat['total'] = 1
df_cat
```

```python
df_cat = df_cat.groupby(['cardio', 'variable','value'], as_index = False).count()
df_cat
```

```python
fig = sns.catplot(x='variable', y='total', col='cardio', data=df_cat, hue='value', kind='bar').fig
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

### E- Relationship between the columns:

Let's draw the heat map, but first we have to clean the data under and above the 2.5 and 97.5 percentile:

```python
df_heat = df[
  (df['ap_lo'] <= df['ap_hi']) &
  (df['height'] >= df['height'].quantile(0.025)) &
  (df['height'] <= df['height'].quantile(0.925)) &
  (df['weight'] <= df['weight'].quantile(0.925)) &
  (df['weight'] >= df['weight'].quantile(0.025))]
df_heat.describe()
```

Let's Calculate the correlation matrix using the pearson method:

```python
corr = df_heat.corr(method='pearson')
corr
```

```python
mask = np.triu(corr)
# Set up the matplotlib figure
fig, ax = plt.subplots(figsize=(12,12))

# Draw the heatmap with 'sns.heatmap()'
sns.heatmap(corr, linewidths=1, annot = True, square = True, mask=mask, fmt='0.1f', center=0.08, cbar_kws={'shrink':0.5})
```
